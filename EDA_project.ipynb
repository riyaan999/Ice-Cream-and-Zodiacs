```python
grouped = df.groupby(['Gender', 'Favourite ice cream flavour?']).size().unstack(fill_value=0)

# Plot the data
grouped.plot(kind='bar', stacked=True, figsize=(8, 6))

# Customize the plot
plt.title("Ice Cream Flavor Preference by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Ice Cream Flavor")
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group the data by gender and ice cream flavor
grouped = df.groupby(['Gender', 'Favourite ice cream flavour?']).size().unstack(fill_value=0)

# Define color palette for the flavors
colors = ["#FF9999", "#66B2FF", "#99FF99", "#FFCC99", "#c2c2f0", "#ffb3e6", "#c2f0c2", "#ffb366"]

# Plot the data with customized styling
ax = grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=colors, width=0.8)

# Customize the plot
plt.title("Ice Cream Flavor Preference by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Ice Cream Flavor", loc='upper left')

# Add labels for each flavor count within the bars
for i, flavor in enumerate(grouped.columns):
    for j, count in enumerate(grouped[flavor]):
        ax.annotate(count, (i, grouped.iloc[:j, i].sum() + count / 2), ha='center', color='black', fontsize=10)

# Beautify the plot further
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(range(len(grouped.index)), grouped.index, rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for age and topping preference
age_groups = ['0-10','10-15','15-20','20-25','25-30','30-35','35+' ]
topping_preferences=df['']

# Generate random counts for each combination of age and topping preference
counts = np.random.randint(1, 11, size=(len(age_groups), len(topping_preferences)))

# Create a heatmap
plt.figure(figsize=(10, 6))
im = plt.imshow(counts, cmap='YlGnBu', aspect='auto', interpolation='none')

# Customize the plot
plt.title('Topping Preference by Age Group')
plt.xlabel('Topping Preference')
plt.ylabel('Age Group')
plt.xticks(np.arange(len(topping_preferences)), topping_preferences, rotation=45)
plt.yticks(np.arange(len(age_groups)), age_groups)
cbar = plt.colorbar(im)
cbar.set_label('Count')

plt.show()
```

```python
ice_cream_counts = df['Favourite ice cream flavour?'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(ice_cream_counts, labels=ice_cream_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Favorite Ice Cream Flavors")
plt.show()
```
